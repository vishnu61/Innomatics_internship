```python
import numpy as np 
import pandas as pd 
import re 
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.iplt20.com/stats/2017/most-runs'
code=requests.get(url)
html=code.text
soup=BeautifulSoup(html)
print(html)
```

```python

Player=[]
url='https://www.iplt20.com/stats/2017/most-runs'
code=requests.get(url)
html=code.text
soup=BeautifulSoup(html)
pos=soup.find_all('div',class_='top-players__player-info')
for i in pos:
    Player.append(i.text)
```

```python
print(len(Player))
```

```python
Mat=[]
soup=BeautifulSoup(html)
mat=soup.find_all('td',class_='top-players__m top-players__padded')
for i in mat:
    Mat.append(i.text)
```

```python
print(len(Mat))
```

```python
Inns=[]  
soup=BeautifulSoup(html)
inn=soup.find_all('td',class_='top-players__inns')
for i in inn:
    Inns.append(i.text)
```

```python
print(len(Inns))
```

```python
NO=[]
soup=BeautifulSoup(html)
no=soup.find_all('td',class_='top-players__no')
for i in no:
    NO.append(i.text)
```

```python
print(len(NO))
```

```python
HS=[]
soup=BeautifulSoup(html)
hs=soup.find_all('td',class_='top-players__hs')
for i in hs:
    HS.append(i.text)
```

```python
print(len(HS))
```

```python
Avg=[]
soup=BeautifulSoup(html)
av=soup.find_all('td',class_='top-players__a')
for i in av:
    if(i=='-'):
        Avg.append(np.NaN)
    else:
        Avg.append(i.text)
```

```python
print(len(Avg))
```

```python
Runs=[]
soup=BeautifulSoup(html)
ru=soup.find_all('td',class_='top-players__r')
for i in ru:
    Runs.append(i.text)
```

```python
print(len(Runs))
```

```python
SR=[]
soup=BeautifulSoup(html)
sr=soup.find_all('td',class_='top-players__sr')
for i in sr:
    SR.append(i.text)
```

```python
print(len(SR))
```

```python
N_4s=[]
soup=BeautifulSoup(html)
s4=soup.find_all('td',class_='top-players__4s')
for i in s4:
    N_4s.append(i.text)
```

```python
print(len(N_4s))
```

```python
N_6s=[]
soup=BeautifulSoup(html)
s6=soup.find_all('td',class_='top-players__6s')
for i in s6:
    N_6s.append(i.text)
```

```python
print(len(N_6s))
```

```python
Year=[]
for i in range(1,101):
    Year.append(2017)
```

```python
df=pd.DataFrame({'Player':Player,'Matches':Mat,'Runs':Runs,'Inns':Inns,'NO':NO,'HS':HS,'Avg':Avg,'SR':SR,'N_4s':N_4s,'N_6s':N_6s,'Year':Year})
df.head()
```

```python
df.Matches=df.Matches.apply(lambda soup:soup.replace('\n',''))
df.Runs=df.Runs.apply(lambda soup:soup.replace('\n',''))
df.Inns=df.Inns.apply(lambda soup:soup.replace('\n',''))
df.NO=df.NO.apply(lambda soup:soup.replace('\n',''))
df.HS=df.HS.apply(lambda soup:soup.replace('\n','').replace('*',''))
df.Avg=df.Avg.apply(lambda soup:soup.replace('\n',''))
df.SR=df.SR.apply(lambda soup:soup.replace('\n',''))
df.N_4s=df.N_4s.apply(lambda soup:soup.replace('\n',''))
df.N_6s=df.N_6s.apply(lambda soup:soup.replace('\n',''))
df.head()
```

```python

df.Avg=df.Avg.apply(lambda soup:soup.replace('\n','')).astype(str)
```

```python
rex=r'\w'
df['Player']=df['Player'].apply(lambda soup:re.compile(rex).findall(soup))
```

```python
df.head()
```

```python
df.Player=df.Player.apply(lambda soup:''.join(soup))
```

```python
df.to_csv('ipl_t20_6.csv',index=False)
```
